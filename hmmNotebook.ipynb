```python
from HiddenMarkovModel.hmmLearnAlgorithm import idealPrediction, predictAndScore, predictNext, averagePredictions
from HiddenMarkovModel.splitData import splitSet, formatData
import numpy as np
import os
```

```python
lengths = []
print(os.getcwd())
print(os.listdir())
```

```python
splitSet("HMM_train_data.csv", 0.1)
observations, lengths = formatData("HMM_training_set.csv")
squishedObservations = np.concatenate(observations)
```

```python
valObservations, valOutcomes = formatData("HMM_validation_set.csv", validation=True)
```

```python
model1 = idealPrediction(squishedObservations, lengths, 30)
```

```python
numNotZero = 0
for observation in observations:
    prediction = averagePredictions(model1, observation, 5, 5)
    if prediction.sum() > 0 : 
        print(prediction)
        numNotZero += 1

numNotZero
```

```python
predictions, nextScore, lastScore = predictAndScore(model1, valObservations, valOutcomes, steps = 3, score = True)
nextScore, lastScore
```
